```python
# 행과 열을 뒤집어주기 : 전치 행렬
a = np.array([[1, 3, 5],
             [7, 9, 11]])

b = np.array([[2, 4],
              [6, 8],
              [10, 12]])

print(a.T + b)
```

```python
a = np.array([[1, 3, 5],
             [7, 9, 11]])

b = np.array([[2, 4],
              [6, 8],
              [10, 12]])
print(a.transpose() + b)
```

```python
a = np.arange(1, 17)
b = a.reshape(2, 2, -1)
c = b.flatten()
d = b.ravel()

print(c)
print(c.base)
print(c.base in c)
print()
print(d)
print(d.base)
print(d.base in d)
```

```python
# 배열 연산
%%time
a = np.arange(1, 10001)
b = np.arange(20001, 30001)

c = np.zeros_like(a)
for i in range(10000):
    c[i] = a[i] + b[i]
```
