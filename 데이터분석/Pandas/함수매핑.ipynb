```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]

def missing_value(n):
    return n.isnull()

result = df.apply(missing_value, axis=1)
result.head()

```

```python
def max_min(n):
    return n.max() - n.min()

result = df.apply(max_min, axis=1)
result
```

```python
def missing_value(n):
    return n.isnull()

def missing_count(n):
    return missing_value(n).sum()

result = df.pipe(missing_count)
result
```
