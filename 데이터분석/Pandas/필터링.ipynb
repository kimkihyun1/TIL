```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

mask1 = (titanic.age >= 10) & (titanic.age < 20)
titanic_teenage = titanic.loc[mask1]
titanic_teenage
```

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
titanic_female_under10 = titanic.loc[mask2, :]
titanic_female_under10
```

```python
mask3 = (titanic.age < 10) | (titanic.age >= 60)
titanic_10_60 = titanic.loc[mask3]
titanic_10_60
```

```python
mask4 = (titanic.sibsp == 3)
mask5 = (titanic.sibsp == 4)
mask6 = (titanic.sibsp == 5)

titanic_boolean = titanic[mask4 | mask5 | mask6]
titanic.head()
```

```python
isin_filter = titanic.sibsp.isin([3, 4, 5])
isin_filter
```

```python
titanic_isin = titanic[isin_filter]
titanic_isin.head()
```
