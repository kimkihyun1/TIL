```python
# 멀티인덱스

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

grouped = df.groupby(['class', 'sex'])

gdf = grouped.mean()
print(gdf.head())
```

```python
print(gdf.loc['First'])
```

```python
print(gdf.loc['First', 'female'])
```

```python
gdf.xs('female', level='sex')
```

```python
# 피벗테이블

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

pdf1 = pd.pivot_table(df,
                      index='class',
                      columns='sex',
                      values='age',
                      aggfunc=['mean', 'sum']
                      )

pdf1
```

```python
pdf2 = pd.pivot_table(df,
                      index=['class', 'sex'],
                      columns='survived',
                      values='age',
                      aggfunc=['mean', 'max']
                      )

pdf2
```

```python
pdf3 = pd.pivot_table(df,
                      index=['class', 'sex'],
                      columns='survived',
                      values=['age', 'fare'],
                      aggfunc=['mean', 'max']
                      )

pdf3
```

```python
pdf3.index
```

```python
pdf3.columns
```

```python
pdf4 = pd.pivot_table(df,
                      index=['class', 'sex'],
                      columns='survived',
                      values=['age', 'fare'],
                      aggfunc=['mean', 'max']
                      )

pdf4.xs('First', level='class')
```

```python
pdf4.xs('First', level=0)
```

```python
pdf4.xs(('First', 'female'), level=('class', 'sex'))
```

```python
pdf4.xs(('First', 'female'), level=(0, 1))
```

```python
pdf4.xs('First')
```

```python
pdf4.xs(('First', 'female'))
```

```python
pdf4.xs('mean', axis=1)
```

```python
pdf4.xs(('mean', 'age'), axis=1)
```

```python
pdf4.xs(('mean', 'age', 1), axis=1)
```

```python
pdf4.xs(1, level='survived', axis=1)
```
