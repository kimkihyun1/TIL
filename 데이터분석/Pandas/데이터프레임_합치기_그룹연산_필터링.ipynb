```python
# 병합 : merge()

pd.set_option('display.max_columns', 10)
pd.set_option('display.max_colwidth', 20)
pd.set_option('display.unicode.east_asian_width', True)

df1 = pd.read_excel('/content/stockprice.xlsx')
df2 = pd.read_excel('/content/stockvaluation.xlsx')

print(df1)
print()
print(df2)
```

```python
merge_outer = pd.merge(df1, df2, how='outer', on='id')
merge_outer
```

```python
merge_left = pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
merge_left
```

```python
agg_mean = grouped.age.mean()
agg_std = grouped.age.std()

agg_mean
```

```python
for key, group in grouped:
    group_zscore = (group.age - agg_mean.loc[key]) / agg_std.loc[key]
    print('key :', key)
    print(group_zscore)
    print()
```

```python
def z_score(x):
    return (x - x.mean()) / x.std()

grouped_age = grouped.age.transform(z_score)
grouped_age
```

```python
grouped_age = grouped.age.transform(lambda x: (x - x.mean()) / x. std())
grouped_age
```

```python
# filter()

def graterthan200(x):
    return len(x) >= 200

grouped_filter = grouped.filter(graterthan200)
grouped_filter.head(10)
```
